```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

# Dados de treino corrigidos
comprimento_treino = np.array([26, 19, 18, 17, 24, 17, 15, 16, 14, 23, 17, 18, 17, 25, 12, 22], dtype=float)
largura_treino = np.array([1.50, 1.70, 1.80, 1.30, 1.30, 1.20, 1.60, 1.40, 1.45, 1.70, 1.35, 1.90, 1.75, 1.80, 1.25, 1.50], dtype=float)
tipo_treino = np.array([0, 2, 2, 0, 1, 2, 2, 2, 1, 0, 2, 2, 2, 1, 0, 1])  # 0: Branco, 1: Amarelo, 2: Rosa

# Dados de teste
comprimento_teste = np.array([15, 16, 17], dtype=float)
largura_teste = np.array([1.77, 1.87, 1.98], dtype=float)

# Converter os rótulos para one-hot encoding
tipo_treino_one_hot = keras.utils.to_categorical(tipo_treino, num_classes=3)
```

```python
# Normalizar os dados
comprimento_treino_norm = comprimento_treino / np.max(comprimento_treino)
largura_treino_norm = largura_treino / np.max(largura_treino)
comprimento_teste_norm = comprimento_teste / np.max(comprimento_treino)
largura_teste_norm = largura_teste / np.max(largura_treino)

# Criar o modelo
model = keras.Sequential([
    keras.layers.Dense(8, activation='relu', input_shape=(2,)),  # Camada oculta com 8 neurônios
    keras.layers.Dense(3, activation='softmax')  # Camada de saída com 3 neurônios (3 classes)
])

# Compilar o modelo
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Treinar o modelo
model.fit(np.column_stack((comprimento_treino_norm, largura_treino_norm)), tipo_treino_one_hot, epochs=200)
```

```python
# Fazer previsões
previsoes = model.predict(np.column_stack((comprimento_teste_norm, largura_teste_norm)))

# Interpretar as previsões
ipe_a = ["Branco", "Amarelo", "Rosa"][np.argmax(previsoes[0])]
ipe_b = ["Branco", "Amarelo", "Rosa"][np.argmax(previsoes[1])]
ipe_c = ["Branco", "Amarelo", "Rosa"][np.argmax(previsoes[2])]

print(f"Ipê A: {ipe_a}")
print(f"Ipê B: {ipe_b}")
print(f"Ipê C: {ipe_c}")
```
